```python
import orjson as json
import os
from glob import glob
from tqdm import tqdm
```

```python
files = sorted(glob('dedup-text-dataset/*.jsonl'))
```

```python
code_files = [
    'code_instructions_120k.jsonl.requested', 
    'python_evol_instruct_51k.jsonl.requested',
    'unnatural-instructions.jsonl.requested'
]
```

```python
for f in code_files:
    with open(f) as fopen:
        for l in fopen:
            l = json.loads(l)
```

```python
l
```

```python
with open('combine.txt', 'w') as fopen_l:
    
    for f in tqdm(code_files):
        with open(f) as fopen:
            for l in fopen:
                l = json.loads(l)
                if len(l['r']) < 100:
                    continue
                fopen_l.write(l['r'] + '\n')
                
    for f in files:
        i = 0
        with open(f) as fopen:
            for l in tqdm(fopen):
                try:
                    l = json.loads(l).strip()
                    fopen_l.write(l + '\n')
                    i += 1
                    if i >= 1e7:
                        break
                except:
                    pass
```
